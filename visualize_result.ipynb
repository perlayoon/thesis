```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 시각화 설정
%matplotlib
%config InlineBackend.figure_format = 'svg'

plt.rcParams['figure.figsize'] = [10, 5]
plt.ion() # enable the interactive mode
```

```python
df = pd.read_csv('./data/exp_result.csv')
```

```python
df
```

```python
# Multi y-axis 차트 시각화

fig, ax = plt.subplots()
fig.subplots_adjust(right=0.75)

twin1 = ax.twinx()

p1, = ax.plot(df.fs_size, df.mae,  "b-", label="MAE")
p2, = twin1.plot(df.fs_size, df.exec_time, "r-", label="Execution Time (sec.)")

ax.set_xlim(42, 2)
ax.set_ylim(0, df.mae.max())
twin1.set_ylim(0, df.exec_time.max())

ax.set_xlabel("Number of Features")
ax.set_ylabel("MAE")
twin1.set_ylabel("Execution Time (sec.)")

ax.yaxis.label.set_color(p1.get_color())
twin1.yaxis.label.set_color(p2.get_color())

tkw = dict(size=4, width=1.5)
ax.tick_params(axis='y', colors=p1.get_color(), **tkw)
twin1.tick_params(axis='y', colors=p2.get_color(), **tkw)

ax.tick_params(axis='x', **tkw)

ax.legend(handles=[p1, p2], loc='upper center')

plt.show()
```
